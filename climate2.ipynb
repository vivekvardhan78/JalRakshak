````python
def load_stock(file, name):
    df = pd.read_csv(file)

    df.columns = df.columns.str.strip().str.replace(' ', '_')

    df['DATE'] = pd.to_datetime(df['DATE'], format='%Y-%m-%d') # Changed this line
    df = df.sort_values('DATE')

    df['Return'] = df['CLOSE'].pct_change()

    df = df[['DATE', 'CLOSE', 'Return']]
    df['Stock'] = name

    return df.dropna()
````

````python
df1 = load_stock('ADANIPOWER_merged_data.csv', 'ADANI')
df2 = load_stock('NTPC_merged_data.csv', 'NTPC')
df3 = load_stock('TATAPOWER_merged_data.csv', 'TATA')
df4 = load_stock('JSWENERGY_merged_data.csv', 'JSW')
df5 = load_stock('POWERGRID_merged_data.csv', 'PGC')

stocks = pd.concat([df1, df2, df3, df4, df5])
stocks.head()
````

````python
climate = pd.read_csv('Temp24-25.csv')

climate['Date'] = pd.to_datetime(climate['YEAR'].astype(str) + '-' + climate['MO'].astype(str) + '-' + climate['DY'].astype(str))
climate = climate.sort_values('Date')

climate = climate[['Date', 'T2M']]
climate = climate.rename(columns={'T2M': 'Temperature'})
````

````python
climate['CDD'] = np.maximum(climate['Temperature'] - 24, 0)
climate['CDD_7'] = climate['CDD'].rolling(7).mean()
climate['Heatwave'] = (climate['Temperature'] > 35).astype(int)

climate = climate.dropna()
````

````python
stocks.rename(columns={'DATE': 'Date'}, inplace=True)

data = pd.merge(stocks, climate, on='Date')
data = data.dropna()

data.head()
````

````python
import pandas as pd
import numpy as np

def expand_timeseries(df, target_size=10000):
    df = df.copy()

    # Ensure sorted
    df = df.sort_values(by='Date').reset_index(drop=True)

    synthetic_data = []

    while len(df) + len(synthetic_data) < target_size:
        temp = df.copy()

        # Add small noise to 'CLOSE' (if present)
        if 'CLOSE' in temp.columns:
            noise = np.random.normal(0, 0.005, size=len(temp))
            temp['CLOSE'] = temp['CLOSE'] * (1 + noise)

        # Shift dates forward
        last_date = df['Date'].max()
        temp['Date'] = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=len(temp), freq='B')

        synthetic_data.append(temp)

        # Update df reference (so next iteration continues forward)
        df = pd.concat([df, temp]).reset_index(drop=True)

    final_df = pd.concat([df] + synthetic_data).iloc[:target_size]

    return final_df


# 🔥 USE THIS
expanded_df = expand_timeseries(data, target_size=10000)

print("Old size:", len(data))
print("New size:", len(expanded_df))
````

````python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from xgboost import XGBClassifier
import matplotlib.pyplot as plt

# --- Re-upload stock data files to ensure availability (addressing FileNotFoundError) ---
from google.colab import files
uploaded = files.upload()
# -------------------------------------------------------------------------

def expand_timeseries(df, target_size=10000):
    df = df.copy()

    # Ensure sorted
    df = df.sort_values(by='Date').reset_index(drop=True)

    synthetic_data = []

    while len(df) + len(synthetic_data) < target_size:
        temp = df.copy()

        # Add small noise to 'CLOSE' (if present)
        if 'CLOSE' in temp.columns:
            noise = np.random.normal(0, 0.005, size=len(temp))
            temp['CLOSE'] = temp['CLOSE'] * (1 + noise)

        # Shift dates forward
        last_date = df['Date'].max()
        temp['Date'] = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=len(temp), freq='B')

        synthetic_data.append(temp)

        # Update df reference (so next iteration continues forward)
        df = pd.concat([df, temp]).reset_index(drop=True)

    final_df = pd.concat([df] + synthetic_data).iloc[:target_size]

    return final_df

# Define load_and_clean function (from id6fbms-e0_4)
def load_and_clean(file, company_name):
    df = pd.read_csv(file)
    df.columns = df.columns.str.strip()
    df['DATE'] = pd.to_datetime(df['DATE'])
    df['Company'] = company_name
    return df

# Define load_stock function (from KTAwIfC4j90U)
def load_stock(file, name):
    df = pd.read_csv(file)
    df.columns = df.columns.str.strip().str.replace(' ', '_')
    df['DATE'] = pd.to_datetime(df['DATE'], format='%Y-%m-%d') # Changed this line
    df = df.sort_values('DATE')
    df['Return'] = df['CLOSE'].pct_change()
    df = df[['DATE', 'CLOSE', 'Return']]
    df['Stock'] = name
    return df.dropna()

# Load individual company data and concatenate (from DQ3igJ6Xf7-q, 18d61e82, 33fa390b, 83c95468, 992adc72)
powergrid_2024 = load_and_clean("Quote-Equity-POWERGRID-EQ-01-01-2024-31-12-2024.csv", "POWERGRID")
powergrid_2025 = load_and_clean("Quote-Equity-POWERGRID-EQ-01-01-2025-31-12-2025.csv", "POWERGRID")
powergrid = pd.concat([powergrid_2024, powergrid_2025]).sort_values(by='DATE')

adpipower_2024 = load_and_clean("Quote-Equity-ADANIPOWER-EQ-01-01-2024-31-12-2024.csv", "ADANIPOWER")
adpipower_2025 = load_and_clean("Quote-Equity-ADANIPOWER-EQ-01-01-2025-31-12-2025.csv", "ADANIPOWER")
adpipower = pd.concat([adpipower_2024, adpipower_2025]).sort_values(by='DATE')

jswenergy_2024 = load_and_clean("Quote-Equity-JSWENERGY-EQ-01-01-2024-31-12-2024.csv", "JSWENERGY")
jswenergy_2025 = load_and_clean("Quote-Equity-JSWENERGY-EQ-01-01-2025-31-12-2025.csv", "JSWENERGY")
jswenergy = pd.concat([jswenergy_2024, jswenergy_2025]).sort_values(by='DATE')

tatapower_2024 = load_and_clean("Quote-Equity-TATAPOWER-EQ-01-01-2024-31-12-2024.csv", "TATAPOWER")
tatapower_2025 = load_and_clean("Quote-Equity-TATAPOWER-EQ-01-01-2025-31-12-2025.csv", "TATAPOWER")
tatapower = pd.concat([tatapower_2024, tatapower_2025]).sort_values(by='DATE')

ntpc_2024 = load_and_clean("Quote-Equity-NTPC-EQ-01-01-2024-31-12-2024.csv", "NTPC")
ntpc_2025 = load_and_clean("Quote-Equity-NTPC-EQ-01-01-2025-31-12-2025.csv", "NTPC")
ntpc = pd.concat([ntpc_2024, ntpc_2025]).sort_values(by='DATE')

# Save merged dataframes to CSVs, so load_stock can read them.
powergrid.to_csv('POWERGRID_merged_data.csv', index=False)
adpipower.to_csv('ADANIPOWER_merged_data.csv', index=False)
jswenergy.to_csv('JSWENERGY_merged_data.csv', index=False)
tatapower.to_csv('TATAPOWER_merged_data.csv', index=False)
ntpc.to_csv('NTPC_merged_data.csv', index=False)

# Load stocks (from gh9Q_nIfkAYX)
df1 = load_stock('ADANIPOWER_merged_data.csv', 'ADANI')
df2 = load_stock('NTPC_merged_data.csv', 'NTPC')
df3 = load_stock('TATAPOWER_merged_data.csv', 'TATA')
df4 = load_stock('JSWENERGY_merged_data.csv', 'JSW')
df5 = load_stock('POWERGRID_merged_data.csv', 'PGC')
stocks = pd.concat([df1, df2, df3, df4, df5])

# Rename 'DATE' column to 'Date' in stocks to match climate DataFrame
stocks.rename(columns={'DATE': 'Date'}, inplace=True)

# Load climate data (from 8-7gW2TwkpY- and yxI1ckQtlALi)
# Corrected filename from Temp24-25.csv to Temp22-25.csv based on kernel state.
climate = pd.read_csv('Temp22-25.csv')
climate['Date'] = pd.to_datetime(climate['YEAR'].astype(str) + '-' + climate['MO'].astype(str) + '-' + climate['DY'].astype(str))
climate = climate.sort_values('Date')
climate = climate[['Date', 'T2M']]
climate = climate.rename(columns={'T2M': 'Temperature'})

# Merge data (from bz6DiZcYlFf8)
data = pd.merge(stocks, climate, on='Date')
data = data.dropna()

# Enhanced Feature Engineering (from m7oT7LcLlo09, hIFYwiUnltla, A6Bu3FY4l-0f, uI9VL9wKmEty, wjdJo9pAmLpU, gXVS2Z6fmNv7)
# --- FIX 1: Lower threshold CDD (more realistic for India) ---
data['CDD'] = np.maximum(data['Temperature'] - 20, 0)

# --- FIX 2: Rolling demand ---
data['CDD_7'] = data.groupby('Stock')['CDD'].rolling(7).mean().reset_index(0, drop=True)

# --- FIX 3: Temperature anomaly ---
data['Temp_Anomaly'] = data['Temperature'] - data['Temperature'].mean()

# --- FIX 4: Dynamic heatwave (top 25% hottest days) ---
threshold = data['Temperature'].quantile(0.75)
data['Heatwave'] = (data['Temperature'] > threshold).astype(int)

# --- FIX 5: Lagged demand (important) ---
data['CDD_Lag1'] = data.groupby('Stock')['CDD'].shift(1)
data['CDD_Lag3'] = data.groupby('Stock')['CDD'].shift(3)

# Ensure data is sorted for lagged features
data = data.sort_values(['Stock', 'Date'])

data['Return_Lag1'] = data.groupby('Stock')['Return'].shift(1)
data['Return_Lag3'] = data.groupby('Stock')['Return'].shift(3)
data['Return_Lag5'] = data.groupby('Stock')['Return'].shift(5)

# Momentum
data['MA_5'] = data.groupby('Stock')['Return'].rolling(5).mean().reset_index(0, drop=True)

# Volatility
data['Volatility'] = data.groupby('Stock')['Return'].rolling(5).std().reset_index(0, drop=True)

# Additional features from later cells
data['Demand_Trend'] = data.groupby('Stock')['CDD'].rolling(3).mean().reset_index(0, drop=True)
data['Trend'] = data['Return_Lag1'] + data['Return_Lag3']
data['Volatility_Change'] = data.groupby('Stock')['Volatility'].diff()

# Drop any NaNs resulting from new feature creation (from dVOgNlf7lveL)
data = data.dropna()

# Re-expand the timeseries to 10,000 records
expanded_df = expand_timeseries(data, target_size=10000)

print("Old size:", len(data))
print("New size:", len(expanded_df))

# REMOVE THE DATE FILTER TO USE THE FULL EXPANDED DATASET
# expanded_df = expanded_df[
#     expanded_df['Date'] <= pd.to_datetime('2026-12-31')
# ]

# Clip outliers for 'Return' columns
expanded_df['Return'] = expanded_df['Return'].clip(-0.2, 0.2)
expanded_df['Return_Lag1'] = expanded_df['Return_Lag1'].clip(-0.2, 0.2)
expanded_df['Return_Lag3'] = expanded_df['Return_Lag3'].clip(-0.2, 0.2)
expanded_df['Return_Lag5'] = expanded_df['Return_Lag5'].clip(-0.2, 0.2)

# Fill any remaining NaNs
expanded_df.fillna(method='ffill', inplace=True)
expanded_df.fillna(method='bfill', inplace=True)

# Recalculate Volatility (after clipping and filling)
expanded_df['Volatility'] = (
    expanded_df.groupby('Stock')['Return']
    .rolling(7).std()
    .reset_index(0, drop=True)
)

# Re-create df from expanded_df and drop NaNs
df = expanded_df.copy()
df = df.sort_values(by=['Stock', 'Date']).reset_index(drop=True)
df = df.dropna()

# Create the 'Direction' column *before* splitting into train/test
df['Direction'] = (df['Return'] > 0).astype(int)

print("\nFinal DataFrame description after reprocessing:")
print(df.describe())
print("\nMissing values after reprocessing:")
print(df.isnull().sum())

# Redefine features and target (if they were changed in the last user instruction)
features = [
    'Temperature', 'CDD', 'CDD_7', 'Heatwave',
    'Return_Lag1', 'Return_Lag3', 'Return_Lag5',
    'Temp_Anomaly', 'CDD_Lag1', 'CDD_Lag3',
    'MA_5', 'Volatility', 'Demand_Trend', 'Trend',
    'Volatility_Change'
]
target = 'Direction'

# Re-split the data
split_date = '2025-06-01'

train = df[df['Date'] < split_date]
test = df[df['Date'] >= split_date]

X_train = train[features]
y_train = train[target]

X_test = test[features]
y_test = test[target]

print(f"\nTrain set size: {len(X_train)}")
print(f"Test set size: {len(X_test)}")

# Removed model training and evaluation from this block, will be in a new cell.
````
